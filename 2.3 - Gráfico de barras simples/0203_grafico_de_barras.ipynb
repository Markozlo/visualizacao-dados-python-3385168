# Gráfico de barras simples

Aprenda a construir gráfico de barras utilizando Matplotlib.


Utilizando um conjunto de dados, modificados para fins didáticos, sobre filmes indicados e ganhadores do Oscar em 2022.

Fonte original dos dados: [Kaggle](https://www.kaggle.com/datasets/pushpakhinglaspure/oscar-dataset?resource=download)


```python
# Importa bibliotecas

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Realiza leitura dos dados

df = pd.read_csv('0203_grafico_de_barras_simples.csv')
```

```python
df.sample(5)
```

```python
# Ordena em ordem decrescente a coluna 'Nomination'

df.sort_values(by='Nomination', ascending=True, inplace=True)
```

```python
# Imprime na tela as cinco primeiras linhas

df.head()
```

```python
# Cria lista com os títulos dos filmes

filmes = df['Film'].to_list()
```

```python
# Cria lista com total de indicações

total_indicacao = df['Nomination'].to_list()
```

```python
# Cria lista com total de prêmios

total_premio = df['Award'].to_list()
```

```python
# Cria visualização

fig, ax = plt.subplots()
ax.bar(filmes, total_indicacao)
plt.show()
```

## Customização

### Etiquetas: rotação

```python
fig, ax = plt.subplots()
ax.bar(filmes, total_indicacao)
plt.xticks(rotation = 90)
plt.show()
```

### Título e etiqueta de eixo

```python
fig, ax = plt.subplots()
ax.bar(filmes, total_indicacao)
plt.xticks(rotation = 90)

ax.set_ylabel('Total de indicações')
ax.set_title('Indicados para o Oscar 2022')

plt.show()
```

### Etiqueta por barra

```python
# Insere etiqueta em cada barra

bar_colors = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#00FFFF', '#FF00FF', '#FFA500', '#800080', '#90EE90', '#ADD8E6', '#FFC0CB', '#A52A2A', '#808080']

fig, ax = plt.subplots()
bar_container = ax.bar(filmes, total_indicacao, color=bar_colors) # Encapsula a configuração do eixo x
ax.bar_label(bar_container) # Adiciona a etiqueta
ax.set_ylabel('Total de indicações')
ax.set_title('Indicados para o Oscar 2022')

plt.xticks(rotation = 90)

plt.show()
```

### Cor

```python
# Estabelece cor para cada barra

bar_colors = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#00FFFF', '#FF00FF', '#FFA500', '#800080', '#90EE90', '#ADD8E6', '#FFC0CB', '#A52A2A', '#808080']

fig, ax = plt.subplots(figsize=(20,7)) # Altera tamanho do quadro
ax.bar(filmes, total_indicacao, color=bar_colors)
plt.xticks(rotation = 90)

ax.set_ylabel('Total de indicações')
ax.set_title('Indicados para o Oscar 2022', fontsize=24) # Altera tamanho da fonte

plt.savefig('grafico_barras_slide.png')
```

### Orientação das barras

```python
bar_colors = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#00FFFF', '#FF00FF', '#FFA500', '#800080', '#90EE90', '#ADD8E6', '#FFC0CB', '#A52A2A', '#808080']

fig, ax = plt.subplots()
bar_container = ax.barh(filmes, total_indicacao, color=bar_colors) # Cria gráfico e barras horizontais
ax.bar_label(bar_container) # Adiciona a etiqueta
ax.set_xlabel('Total de indicações')
ax.set_title('Indicados para o Oscar 2022')

plt.show()
```
